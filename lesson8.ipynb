```python
def simulate_hat_game(num_cats, num_rounds):
    cats = [False] * num_cats # Початково всі коти без капелюхів
    
    for round_number in range(1, num_rounds + 1): 
        # Цей цикл обходить кожен кіт згідно індексу, враховуючи підходи для раундів
        for cat_index in range(num_cats):
            if round_number == 1 or (round_number % 2 == 0 and cat_index % 2 == 1) or (round_number % 3 == 0 and cat_index % 3 == 2):
                cats[cat_index] = not cats[cat_index]
                
    # Виведення результатів
    for cat_number, has_hat in enumerate(cats, start=1):
        print(f"Кіт №{cat_number}: {'Є капелюх' if has_hat else 'Немає капелюха'}")
        
# Робимо 100 обходів котів
simulate_hat_game(num_cats=100, num_rounds=100) 
```
